```python
# import the requred library
from bs4 import BeautifulSoup
# import urllib2 ; instead of this execute the below code
from urllib.request import urlopen
```

```python
# url for web scrapping
url = 'https://simplilearn.com/resources'
```

```python
# Scrape the webpage document
# webpage = urllib2.urlopen(url) # change this step to below one
webpage = urlopen(url)
```

```python
# create a soup object and parse the webpage using html parser
sl_soup = BeautifulSoup(webpage, "html.parser")
```

```python
# close the webpage object
webpage.close()
```

```python
# View the content of the soup object
sl_soup.contents
```

```python
# use the prettify method to view well formatted html output
print(sl_soup.prettify())
```

```python
# view the title of the scrapped document
sl_soup.title
```

```python
# find all href link present in webpage
for href in sl_soup.find_all('a', href=True):
    print(href['href'])
```

```python
# view all the headers name
for article in sl_soup.find_all("h2"):
    print(article.string)
```

```python
for article_topic in sl_soup.find_all("h4"):
    print(article_topic.string)
```

```python
url2 = "https://www.simplilearn.com/what-is-agile-project-management-article?source=frs_home"
webpage2 = urlopen(url2)
sl_article = BeautifulSoup(webpage2, "html.parser")
test = sl_article.find(class_="desig_author empty-text")
```

```python
# view the type of the article object
type(test)
```

```python
# find first article name
first_next_p = test.p
first_next_p
```

```python
# find next siblings (next article name)
first_next_sibling = first_next_p.next_sibling.next_sibling
first_next_sibling
```

```python
# find previous sibling(previous article name)
first_next_sibling.previous_sibling.previous_sibling
```
