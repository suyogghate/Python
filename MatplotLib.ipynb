```python
# Use nukpy to generate random data
import numpy as np

x = np.linspace(0, 10, 25)
y = x * x + 2
print(x)
print(y)
```

```python
# It only one command to plot graph
pylab.plot(x,y, 'g')  # 'g' stands for green
```

```python
# Drawing a subgraph
pylab.subplot(1, 2, 1) # The contents of the brackets represents (row, column, index)
pylab.plot(x, y, 'r--') # third parameter here is color and line style
pylab.subplot(1, 2, 2)
pylab.plot(y, x, 'g*-')
# play around with these numbers(r, c, i)
# (1, 2, 1) and (1, 2, 2) are significant if we want tiled subplots, without whch both
# graphs will be in the same plot.
```

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()
axis = fig.add_axes([0.5, 0.1, 0.8, 0.8])  # Controls the left, right, width, height of the canvas (from 0 to 1)
axis.plot(x, y, 'r')
```

```python
# again we can draw subgraphs
fig, axes = plt.subplots(nrows=1, ncols=2)  # subap is of 1 row, 2 cols
for ax in axes:
    ax.plot(x, y, 'r')
```

```python
# we can also draw a picture, or graph inside another graph
fig = plt.figure()

# Control the left, bottom, width, height of the canvas

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # big axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # small axes

axes1.plot(x, y, 'r')
axes2.plot(x, y, 'g')
```

# Other 2D Graphs

```python
n = np.array([0, 1, 2, 3, 4, 5])

fig, axes = plt.subplots(1, 4, figsize=(16, 5))
axes[0].set_title("scatter")
axes[0].scatter(x, x + 0.25*np.random.randn(len(x)))

axes[1].set_title("step")
axes[1].step(n, n**2, lw=2)

axes[2].set_title("bar")
axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)

axes[3].set_title("fill_between")
axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5)
```

```python
# Draw a radar chart
fig = plt.figure(figsize=(6, 6))
ax = fig.add_axes([0.0, 0.0, 1, 1], polar=True)
t = np.linspace(0, 2 * np.pi, 100)
ax.plot(t, t*2, color='blue', lw=3)
```

```python
# Draw a histogram
n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].set_title("Default histogram")
axes[0].hist(n)

axes[1].set_title("Cumulative detailed histogram")
axes[1].hist(n, cumulative=True, bins=10)
```

```python
# Draw a contour image
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

print(X)
print(Y)
print(Z)
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('contour map')
```
